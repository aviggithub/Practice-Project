```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

income = np.random.normal(20,5,30) #normal(loc(mean),variance,total random number)
#below print data show in Array 
#print(income)

#convert array to datafrmae
incomedf=pd.DataFrame(income)
incomedf.columns=['Income']
#print(incomedf)

#print first 5 records
incomedf.head(5)

#show Histogram
plt.hist(incomedf.values,30)
plt.show()

#dataframe to array 
incomeary=incomedf.values

#Standard deviatiion andvariance and mean
#mean = (sum of total num/n)  std/sigma = (x-mean/N)  variance/sigma square = (x-mean)2/N  N=number of values
print("Standard Deviation ",incomedf.std(),"Variance",incomedf.var(), "Mean",incomedf.mean())




```

```python
#Moments and percentile
#moments :Mean,variance,Skew,Kurtosis
import scipy.stats as sp
print(sp.skew(income))
print(sp.kurtosis(income))
```

```python
from datetime import datetime
from nsepy import get_history
import datetime
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import Series
from pandas import concat
from pandas import Series
from pandas import DataFrame
from pandas.tools.plotting import lag_plot
from pandas.tools.plotting import autocorrelation_plot
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.ar_model import AR
from sklearn.metrics import mean_squared_error

start12 =datetime.datetime.strptime(input("Enter the stock price Start date in DD-MON-YYYY :"),"%d-%b-%Y")
end12 =datetime.datetime.strptime(input("Enter the stock price End date in DD-MON-YYYY :"),"%d-%b-%Y")
stock12 =input("Enter the stock code e.g.: Tata Motors : TATAMOT   :")
# Extrating data
ext_stock122 = get_history(stock12,start12,end12)
ndf=pd.DataFrame(ext_stock122)
ndf.head()
#Moving Average.
mov_avg = ndf.copy()
mov_avg['moving_avg_forecast'] = ndf['Close'].rolling(20).mean().iloc[-1]
plt.figure(figsize=(16,8))
#plt.plot(train['Close'], label='Train')
plt.plot(ndf['Close'], label='Train')
#plt.plot(test['Close'], label='Test')
plt.plot(mov_avg['moving_avg_forecast'], label='Moving Average Forecast')
plt.legend(loc='best')
plt.show()

rmsmv = np.sqrt(mean_squared_error(ndf.Close, mov_avg.moving_avg_forecast))
print("RMSE=",rmsmv)
```

```python
from statsmodels.tsa.api import ExponentialSmoothing, SimpleExpSmoothing, Holt
import numpy as npp
fit2 = SimpleExpSmoothing(npp.asarray(ndf['Close'])).fit(smoothing_level=0.6,optimized=False)
exp_smoothm['SES'] = fit2.forecast(49)
plt.figure(figsize=(16,8))
plt.plot(ndf['Close'], label='Train')
plt.plot(exp_smoothm['SES'], label='SES')
plt.legend(loc='best')
plt.show()
rmssme = npp.sqrt(mean_squared_error(ndf.Close, exp_smoothm.SES))
print("RMSE=",rmssme)
```
